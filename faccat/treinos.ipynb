```python
# Ler um número e saber se ele é par ou ímpar
numero = int(input())

if numero % 2 == 0:
    print(f'Este número é par')
else:
    print(f'Este número é ímpar')
```

```python
n1, n2, n3 = map(int, input().split())

if n1 > n2 and n1 > n3:
    print(f'O maior número é {n1}')
elif n2 > n3 and n2 > n1:
    print(f'O maior número é {n2}')
else:
    print(f'O maior número é {n3}')
```

```python
num = int(input())

if num > 0:
    print(f'Este número é positivo')
elif num < 0:
    print(f'Este número é negativo')
else:
    print(f'Este número é zero')
```

```python
soma, contagem = 0, 0
while True:
    escolha = int(input())

    if escolha == 0:
        break

    num = int(input())
    soma += num
    contagem += 1

print(f'A soma dos valores é de: {soma}')
print(f'A quantidade de números digitados: {contagem}')
print(f'A média é de: {soma / contagem:.2f}')
```
